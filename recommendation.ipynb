```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import dask.dataframe as dd
```

```python
d1 = pd.read_csv('data.csv')
d1.head()
```

```python
import dask.dataframe as dd

d2 = dd.read_csv('users.csv')
d2.head()
```

```python
d2.tail()
```

```python
d2['title'] = d2['title'].str.slice(0, -7)
d2.head()
```

```python
d2 = d2.rename(columns={'title': 'movie_title'})
d2.head()
```

```python
d2 = d2.drop(['genres', 'timestamp'], axis=1)
d2.head()
```

```python
d1['movie_title'] = d1['movie_title'].str.strip()
d2['movie_title'] = d2['movie_title'].str.strip()
```

```python
merged_df = pd.merge(d1, d2, on='movie_title')
merged_df
```

```python
merged_df.to_csv('recommendations.csv', index=False)
```
